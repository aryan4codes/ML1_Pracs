```python
import pandas as pd

df= pd.read_csv("CarPrice_Assignment.csv")
```

```python
# df.drop(['car_ID']
df.info()
```

```python
df = df.select_dtypes(exclude=['object'])

# Display the resulting DataFrame
df
```

```python
df.drop(columns=['car_ID'],inplace=True)
```

```python
df.nunique()
```

```python
from sklearn.model_selection import train_test_split



X= df.drop('price',axis=1)
y= df['price']


train_X, test_X, train_y, test_y = train_test_split(X,y,test_size = 0.3, random_state=2)

from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()

model.fit(train_X,train_y)
```

```python
y_pred= model.predict(test_X)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score


print("MSE :",mean_squared_error(y_pred,test_y))
print("MAE :",mean_absolute_error(y_pred,test_y))
print("R2 Score",r2_score(y_pred,test_y))
```

```python
train_score, test_score = list(),list()

for i in range(1,21):
    model = model = DecisionTreeRegressor(max_depth=i)
    model.fit(train_X,train_y)
    train_ypred= model.predict(train_X)
    train_acc = r2_score(train_ypred,train_y)
    train_score.append(train_acc)
    
    test_ypred= model.predict(test_X)
    test_acc = r2_score(test_ypred,test_y)
    test_score.append(test_acc)
    
    print("%d Training Score: %.3f \n Testing Score : %.3f \n\n\n"%(i,train_acc,test_acc))
```

```python
import matplotlib.pyplot as plt

plt.plot(train_score,marker='o',label="Training Score")
plt.plot(test_score,marker='o',label="Testing Score")

plt.legend()
```

### After depth of 3 , model starts to overfit
